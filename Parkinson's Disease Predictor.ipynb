**Using SMOTE to oversample the minority class**

There are different method to oversample the minority class.
we can refer <a herf="https://www.analyticsvidhya.com/blog/2020/10/overcoming-class-imbalance-using-smote-techniques/">SMOTE Tutorial</a>

```python
! pip install imblearn
```

```python
# Compare different sampling method
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import ADASYN
from imblearn.combine import SMOTETomek
from imblearn.combine import SMOTEENN
from collections import Counter
```

```python
X, y = parkinsons.drop(['name', 'status'], axis=1), parkinsons['status']
```

```python
# SMOTE
smote = SMOTETomek(random_state=50)
X_smote, y_smote = smote.fit_resample(X, y)

# ADASYN: Adaptive Synthetic Sampling Approach
ada = ADASYN(random_state=100)
X_ada, y_ada = ada.fit_resample(X, y)

# Hybridization: SMOTE + Tomek Links
smote_tomek = SMOTETomek(random_state=250)
X_smote_tomek, y_smote_tomek = smote_tomek.fit_resample(X, y)

# Hybridization: SMOTE + ENN
smote_enn = SMOTEENN(random_state = 300)
X_smote_enn, y_smote_enn = smote_enn.fit_resample(X, y)
```

```python
print('Before sampling', Counter(parkinsons.status))
print('SMOTE samplng', Counter(y_smote))
print('ADASYN samplng', Counter(y_ada))
print('SMOTE + Tomek Links', Counter(y_smote_tomek))
print('SMOTE + ENN', Counter(y_smote_enn))
```

```python
# PCA analysis
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
X_list = [X, X_smote, X_ada, X_smote_tomek, X_smote_enn]
y_list = [y, y_smote, y_ada, y_smote_tomek, y_smote_enn]
methods = ['Before Sampling', 'SMOTE', 'ADASYN', 'SMOTE + Tomek', 'SMOTE + ENN']

import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

sc= StandardScaler()

_, axes = plt.subplots(1, 5, figsize=(24, 4))
i = 0
for XX, yy, m in zip(X_list, y_list, methods):
  # standarize data
  XX_scaled = sc.fit_transform(XX)
  components=pca.fit_transform(XX_scaled)
  df_components=pd.DataFrame(data=components,
                        columns=['PC1','PC2'])
  df_combined = pd.concat([df_components, pd.Series(yy, name='status')], axis=1)
  sns.scatterplot(ax=axes[i], data=df_combined, x='PC1', y='PC2', hue='status')
  axes[i].set_title(m)
  i += 1
```

It looks like that combination of SMOTE and ENN is best to seperate two clusters due to the two clusters have least overlap region.

```python
df_list = [df_original, df_smote, df_ada, df_smote_tomek, df_smote_enn]
methods = ['Before Sampling', 'SMOTE', 'ADASYN', 'SMOTE + Tomek', 'SMOTE + ENN']
```

**Split dataset**, **Data preprocessing**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
import matplotlib.pyplot as plt

_, axes = plt.subplots(1, 5, figsize=(24, 4))
i = 0
for df, m in zip(df_list, methods):
  # Declare features and target
  X, y = df.drop(['status'], axis=1), df['status']

  # Splite dataset
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

  # Standarized data
  sc= StandardScaler()
  X_train_scaled = sc.fit_transform(X_train)
  X_test_scaled = sc.transform(X_test)

  # PCA transform
  pca = PCA(n_components=0.85, svd_solver='full')
  pca.fit(X_train_scaled)
  components = pca.explained_variance_ratio_.cumsum()
  # print(f'Number of componentes {len(components)}')
  axes[i].plot(range(1, len(components)+1), components, 'go--', linewidth=2, markersize=12 )
  axes[i].set_title(f'{m} ({len(components)} components)')
  i += 1
```

If we want to extract 90% variance, there are 5, 4, 5, 4, and 4 components can will be used for the dataset without samping, with SMOTE sampling, with ADASYN sampling, with SMOTE and Tomek sampling, and with SMOTE and ENN sampling

## **Model Construction**

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
_sc = StandardScaler()
_pca = PCA(n_components=0.85)
```

### **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
_model = LogisticRegression()
log_regress_model = Pipeline([
    ('std_scaler', _sc),
    ('pca', _pca),
    ('regressor', _model)
])
```

```python
from sklearn import set_config
set_config(display='diagram')
display(log_regress_model)
```

```python
from sklearn.metrics import classification_report

acc = {}
predictions={}
reports = {}
for df, m in zip(df_list, methods):
  # Declare features and target
  X, y = df.drop(['status'], axis=1), df['status']
  # Splite dataset
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
  log_regress_model.fit(X_train, y_train)

  # Evaluation and prediction
  acc[m]= log_regress_model.score(X_test, y_test)
  y_pred = log_regress_model.predict(X_test)
  predictions[m] = y_pred

  # Classification Report
  reports[m] = classification_report(y_test, y_pred)
```
